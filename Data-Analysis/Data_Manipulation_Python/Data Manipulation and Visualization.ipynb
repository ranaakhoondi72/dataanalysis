```python
from scipy import stats
outlier_threshold = 3

# Initialize a counter for total outliers
total_outliers = 0

# Loop through each column in the DataFrame
for column in df3.columns:
    if df3[column].dtype != 'object':  # Skip non-numeric columns
        # Calculate Z-scores
        z_score = np.abs(stats.zscore(df3[column]))
        
        # Identify outliers for the current column
        outliers_zscore = df3[z_score > outlier_threshold]
        
        # Count and display the number of outliers for this column
        column_outliers = len(outliers_zscore)
        total_outliers += column_outliers
        
        print(f"Total outliers in '{column}': {column_outliers}")
        if column_outliers > 0:
            print(outliers_zscore)  # Optionally print the outlier rows for the column

# Print total number of outliers across all columns
#print("\nTotal outliers across all columns:", total_outliers)
```

```python
from scipy import stats
# detecting outliers for each column and show each column outliers as well as total outliers
outlier_threshold = 3
total_outliers = 0

# make a dictionary to store the count of outliers for each column
outlier_summary = {}

# Loop for each column in the dataframe
for column in df4.columns:
    if df4[column].dtype != 'object':  # just go for numeric column
        # Calculate Z-scores
        z_score = np.abs(stats.zscore(df4[column]))
        
        # find outliers for the current column
        outliers_zscore = df4[z_score > outlier_threshold]
        
        # Count the number of outliers for this column
        column_outliers = len(outliers_zscore)
        total_outliers += column_outliers
        
        # save  count of outliers in  summary dictionary
        outlier_summary[column] = column_outliers

# show the summary of outliers for each column
for column, count in outlier_summary.items():
    print(f"Total outliers in '{column}': {count}")

# show total number of outliers for all data set -> numerical
print("\nTotal outliers across all columns:", total_outliers)
```

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns 
# Extract Year, Month, Day from 'Date'
df4['Year'] = df4['Date'].dt.year
df4['Month'] = df4['Date'].dt.month
df4['Day'] = df4['Date'].dt.day

# Select only the numeric columns for correlation analysis
numeric_data = df4.drop(columns=['Date', 'Name'])  # Exclude non-numeric columns

 #Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

 #Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix of Numerical and Time-Based Features")
plt.show()
```

```python
# Plot time series of each numerical column with the 'Date' 
plt.figure(figsize=(14, 10))

# Choose numerical columns to plot
columns_to_plot = ['Open', 'Close', 'Low', 'High', 'Volume']

# Plot each column in time pass
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(len(columns_to_plot), 1, i)  # Subplot for each variable
    plt.plot(df4['Date'], df4[column], label=column, color='b' if column != 'Volume' else 'g')
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.title(f'{column} Over Time')
    plt.grid(True)
    plt.tight_layout()

plt.show()
```
